```python
import ir_datasets
import json
from collections import defaultdict
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

# تحميل الحزم لمرة واحدة محلياً
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

# تجهيز أدوات التنظيف الصرفي
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

def clean_text(raw_text):
    if not raw_text:
        return []
    tokens = word_tokenize(raw_text.lower())
    return [stemmer.stem(word) for word in tokens if word.isalnum() and word not in stop_words]

print("✅ تم استدعاء المكتبات وتجهيز دالة التنظيف بنجاح!")
```

```python
dataset = ir_datasets.load("msmarco-document/dev")

processed_corpus = {}
LIMIT = 5000  # عينة جيدة جداً للتأكد من صحة الفهرس ومطابقة الشروط مبدئياً

print("⏳ جاري قراءة المستندات وتنظيفها... يرجى الانتظار ثوانٍ...")

for i, doc in enumerate(dataset.docs_iter()):
    if i >= LIMIT:
        break
    
    # تنظيف النص وحفظ النتيجة بداخل الـ Dictionary
    processed_corpus[doc.doc_id] = clean_text(doc.text)
    
    if (i + 1) % 1000 == 0:
        print(f"✔️ تمت معالجة {i + 1} وثيقة...")

print(f"📊 انتهت المعالجة! إجمالي الوثائق الجاهزة في الذاكرة: {len(processed_corpus)}")
```

```python
# طباعة أول وثيقة كمثال لمعاينة شكل البيانات
first_id = list(processed_corpus.keys())[0]
print(f"Document ID: {first_id}")
print(f"Processed Tokens (First 20 words): {processed_corpus[first_id][:20]}")
```
